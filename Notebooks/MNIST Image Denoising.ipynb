## Loading Libraries

```python
import torch
from src.train import train
from src.model import ConvDenoiser
from src.utils import show_images,load_data
from src.evaluate import eval
```

## Loading the data

```python
train_loader, test_loader = load_data()
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model = ConvDenoiser()
```

## Training the model

```python
model = train(model,train_loader=train_loader,device=device,epochs = 200,learning_rate=0.01)
```

```python
denoised_imgs,noisy_test_imgs,test_imgs =  eval(model,test_loader,device)
```

## The results

```python
show_images(denoised_imgs,noisy_test_imgs,test_imgs,n=9)
```
